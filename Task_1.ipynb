```python
## Visualize outliers using boxplots and remove them ##

## Storing the boxplot in a local storage.

import matplotlib.pyplot as plt
folder='titanic_table1'
cwd=os.getcwd()
new_dir=os.path.join(cwd,folder)
try:
    os.makedirs(new_dir)
except Exception as e:
    print(e)

for i in numerical:
    path=f'{i}_boxplot.jpg'
    new_path=os.path.join(new_dir,path)
    plt.boxplot(Titanic_data[i])
    plt.title(f'{i}_boxplot')
    plt.savefig(new_path)
    plt.show()
```

```python
for i in numerical:
    Q1=np.quantile(Titanic_data[i],0.25)
    Q3=np.quantile(Titanic_data[i],0.75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    Median=Titanic_data[i].median()
    cond1=Titanic_data[i]<lb
    cond2=Titanic_data[i]>ub
    cond=cond1|cond2
    true=Median
    false=Titanic_data[i]
    Titanic_data[i]=np.where(cond,true,false)
    
    
    
```
