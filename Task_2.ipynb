```python
## drop the column 
Titanic_data=Titanic_df.drop(labels=['PassengerId'],axis=1)
Titanic_data
```

```python
cat=Titanic_data.select_dtypes(include='object').columns
num=Titanic_data.select_dtypes(exclude='object').columns
cat,num
```

```python
Titanic_data.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in cat:
    Mode=Titanic_data[i].mode()
    Titanic_data[i].fillna(Mode.values[0],inplace=True)
Titanic_data
```

```python
Titanic_data.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in num:
    Median=Titanic_data[i].median()
    Titanic_data[i].fillna(Median,inplace=True)
Titanic_data
```

```python
## 1.Generate summary statistics (mean, median, std, etc.).

#calculate mean
mean_summary = Titanic_data.mean(numeric_only=True)
print("The Value of mean is:", mean_summary)

# calculate median
median_summary =Titanic_data.median(numeric_only=True)
print("The Value of median is:", median_summary)

# calculate standard deviation
std_summary =Titanic_data.std(numeric_only=True)
print("The Value of standard deviation is:", std_summary)

# calculate minimun 
min_summary =Titanic_data.min(numeric_only=True)
print("The Value of minimum is:", min_summary)

# calculate maximum
max_summary =Titanic_data.max(numeric_only=True)
print("The Value of maximum is:", max_summary)
```

```python
numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Create pairplot
sns.pairplot(Titanic_data[numeric_cols])
plt.suptitle("Pairplot of Titanic Numeric Features", y=1)
plt.show()
```

```python
## 4.Identify patterns, trends, or anomalies in the data

########## Patterns & Trends ##########

print("\n--- Survival Rate by Gender ---")
survival_by_gender = Titanic_data.groupby('Sex')['Survived'].mean()
print(survival_by_gender)

print("\n--- Survival Rate by Passenger Class ---")
survival_by_class = Titanic_data.groupby('Pclass')['Survived'].mean()
print(survival_by_class)

print("\n--- Survival Rate by Embarked Port ---")
survival_by_embarked = Titanic_data.groupby('Embarked')['Survived'].mean()
print(survival_by_embarked)

# Plot survival by gender
sns.countplot(data=Titanic_data, x='Sex', hue='Survived')
plt.title("Survival Count by Gender")
plt.show()

# Plot survival by class
sns.countplot(data=Titanic_data, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.show()

# Plot age distribution by survival
sns.histplot(data=Titanic_data, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()

######## Anomalies #########

print("\n--- Missing Values ---")
missing_values = Titanic_data.isnull().sum()
print(missing_values)

# Boxplot to detect fare outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=Titanic_data['Fare'])
plt.title("Fare Outliers")
plt.show()

# Show most common duplicate tickets
print("\n--- Most Common Shared Tickets ---")
ticket_counts = Titanic_data['Ticket'].value_counts()
shared_tickets = ticket_counts[ticket_counts > 1]
print(shared_tickets.head())

########### Correlation Matrix (Numeric Pattern) ############

numeric_cols =['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
correlation_matrix = Titanic_data[numeric_cols].corr()

print("\n--- Correlation Matrix ---")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
## 5.Make basic feature-level inferences from visuals.


# Set plot style
sns.set(style="whitegrid")

## Survival by Gender ##
print("\nðŸ“Œ Inference: Females had a much higher survival rate than males.\n")
sns.countplot(data=Titanic_data, x='Sex', hue='Survived')
plt.title("Survival by Gender")
plt.show()

## Survival by Passenger Class ##
print("\nðŸ“Œ Inference: 1st class passengers were more likely to survive than 2nd and 3rd class passengers.\n")
sns.countplot(data=Titanic_data, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.show()

## Age Distribution by Survival ##
print("\nðŸ“Œ Inference: Children had better survival chances. Elderly had lower survival rates.\n")
sns.histplot(data=Titanic_data, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()

## Fare Distribution (Boxplot) ##
print("\nðŸ“Œ Inference: A few passengers paid very high fares (outliers), likely wealthy 1st class passengers.\n")
sns.boxplot(x=Titanic_data['Fare'])
plt.title("Fare Distribution with Outliers")
plt.show()

## Correlation Matrix ##
print("\nðŸ“Œ Inference: Fare is negatively correlated with Pclass. SibSp and Parch are positively correlated.\n")
numeric_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
corr_matrix = Titanic_data[numeric_features].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```
